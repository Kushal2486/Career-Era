```python
import pandas as pd 

# Here, note that we will set the index to be the id column 
df = pd.read_csv('https://raw.githubusercontent.com/JD-DataScienceProjects/Careerera_Projects/main/AB_NYC_2019.csv', index_col = 'id')

df.head()
```

```python
#3. What is the shape of the data
# The shape of the data can be determined by using shape as shown below

df.shape
```

```python
# 4. How many numerical features are there?
# We can use decribe for this purpose as it deals with numrical values only

df.describe()
```

```python
# 5. How many non-numerical features are there?

# Info helps us understand he data types for each column as well as the null and non null values

df.info()
```

```python
# By using info its clear that there are 6 nin numerical featues. 
# Namely - 'name', 'hist_name', 'neighbourhood_group', 'neighbourhood', 'last_review'
```

```python
# 6. Drop the 'name' and 'host_name' columns

# In order to drop the name and host_name we can use the drop() function.We could have used the inplace=True 
# to delete the columns permanenelty but it seems to affect the questins that come late and hence didnt use.
# I would be grateful if you could shed some light on what would have been an ideal path.

df.drop(['name','host_name'],axis = 1)
df.shape  # We are using the shape command to check the number of columns have reduced or not.
```

```python
# 7. What is the total number of missing values?

# df.isnull() helps to determine the number of null values but to find the total number of null values we must
# use sum() along with isnull() to get the total number of null values in each column.
# We will use tolist() to convert the same into list and use the sum() to find the total of the list

null_values_list = df.isnull().sum().tolist()  
sum(null_values_list)
```

```python
#8. Which columns have the most missing values?

# df.isnull() helps to determine the number of null values but to find the total number of null values we must
# use sum() along with isnull() to get the total

df.isnull().sum()  # Its clear that 'last_reviews' and 'review_per_month' have the most number of missing values
```

```python
# 9. What kind of variable are the columns with the most missing values?

df.info()
```

```python
# df.info() tell us that 'last_reviews' and 'review_per_month' are object and float64 variable type respectively
```
